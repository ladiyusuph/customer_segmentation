```python
#Finding wcss values for different number of clusters

wcss = []
SEED = 234
for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = SEED)
    kmeans.fit(X)
    
    wcss.append(kmeans.inertia_)
```

```python
#Plotting the Elbow Graph
sns.set()
plt.plot(range(1,11),wcss)
plt.title('Elbow Graph')
plt.ylabel('WCSS')
plt.xlabel('n_clusters')
plt.show()
```

```python
#Training the clustering algorithm
kmeans = KMeans(n_clusters = 5, init = 'k-means++', random_state = SEED)


#Return a label for each data point based on their clusters
y1 = kmeans.fit_predict(X)

print(y1)
```

```python
# plotting all the clusters and their Centroids

plt.figure(figsize=(6,7))
plt.scatter(X[y1==0,0], X[y1==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[y1==1,0], X[y1==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[y1==2,0], X[y1==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[y1==3,0], X[y1==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[y1==4,0], X[y1==4,1], s=50, c='blue', label='Cluster 5')

# plot the centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s=100, c='cyan', label='Centroids')

plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```
